```python
import seaborn.objects as so
import pandas as pd

# GET DATA
URL = 'https://learn.sharpsightlabs.com/datasets/pdm/supercars.csv'
supercars = pd.read_csv(URL)

# MAKE SCATTERPLOT WITH TRENDLINE
(so.Plot(data = supercars
         ,x = 'horsepower'
         ,y = 'torque')
  .add(so.Dots(color = 'navy'))                           #add scatter dots
  .add(so.Line(color = 'orange'), so.PolyFit(order = 1))  #add trend line
)

#If you use `so.Line` all by itself, it will try to connect all of the points with a single line.
#To turn this into a trend line, we need to use `so.PolyFit`.
#And to turn this specifically into a linear trend line, we need to use `order = 1`.
```
